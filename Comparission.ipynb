```python
# lets practice on comparission operators
Nisha =80 #she got 80% of score in semester
Agarwal =95 #she also scored 95% of score in semester
# lets check who got more score
print(Nisha>Agarwal)
print("Nisha got less marks than Agarwal")
print(Agarwal>Nisha)
print("Agarwal got more marks than Nisha")
```

```python
# Here in another case we check
Nisha =95 #in semester 2 she scored 95%
Agarwal =95 #in semester 2 she again scored same percentage 95%
print(Nisha == Agarwal)
print("At this time Nisha was equally scored to Agarwal")
print("Congrats Nisha")
```
